```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_results = pd.read_csv('results.csv')
filters = ['F555W','F814W','F090W','F150W','F160W','F277W']
for filt in filters:
    dmag = df_results[filt+'_mag_sphot'] - df_results[filt+'_mag_ceers']
    df_results[filt+'_dmag'] = dmag

print(df_results.shape)
df_results.head()
```

```python
plt.scatter(df_results['aper_size'],df_results['F555W_dmag'])
```

```python
plt.scatter(df_results['aper_size'],df_results['F150W_mag_ceers'])
```

```python
plt.scatter(df_results['F555W_dmag'],df_results['F277W_dmag'])
plt.xlim(-1,1)
plt.ylim(-1,1)
```

```python
filt = 'F277W'
xdata = df_results[filt+'_mag_ceers']
ydata = df_results[filt+'_mag_sphot']
yerr = df_results[filt+'_err_sphot']
plt.errorbar(xdata,ydata,yerr=yerr,fmt='o')
plt.plot([xdata.min()-0.5,xdata.max()+0.5],[xdata.min()-0.5,xdata.max()+0.5],'k--')
```
